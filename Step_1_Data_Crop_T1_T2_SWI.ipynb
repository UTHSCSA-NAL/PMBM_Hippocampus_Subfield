```python
import nibabel as nib
import numpy as np
```

```python
name =["21-r", "22", "23-r", "24-r", "28-r", "29-r", "35", "36", "37-r", "38-1", "39",  "40", "42",  "45", "48"]
location= ["127:223,212:260", "147:243,162:210", "142:238,167:215", "130:226,212:260", "130:226,205:253", "130:226,210:258", "130:226,210:258", "127:223,200:248", "127:223,210:258", "127:223,210:258", "137:233,180:228",  "127:223,210:258", "127:223,190:238",  "127:223,210:258", "127:223,200:248"]
for i in range(0,15):
    print(location[i], name[i])
   
    #name = "48"
    Data_crop=nib.load('/media/abn/Hard_Disk/Registration/Data_T1/PMBM'+ str(name[i])+'_Rigid_reg_t1_to_SWI.nii.gz')
    Data1_crop = Data_crop.get_fdata()
    print(Data1_crop.shape)
    import cv2
    import matplotlib.pyplot as plt
    crop_image = np.zeros((Data1_crop.shape[0],96,48))
    for loop in range (0,Data1_crop.shape[0]):
        Slice = Data1_crop[loop,:,:]
        l = location[i]
        #print(l)
        l1 = int(l[:3])
        l2 = int(l[4:7])
        l3 = int(l[8:11])
        l4 = int(l[12:15])
        crop_img = Slice[l1:l2,l3:l4]
        crop_image[loop,:,:] = crop_img
        if np.max(Slice) == 1:
            plt.imshow(Slice*255)
            plt.show()
            plt.imshow(crop_img*255)
            plt.show()
    crop_image = crop_image[62:126,:,:]
    final_img = nib.Nifti1Image(crop_image, Data_crop.affine)
    
    nib.save(final_img, '/media/abn/Hard_Disk/Registration/Data_T1/Data_cropped_T1/PMBM'+ str(name[i])+'_Rigid_reg_t1_to_SWI_cropped.nii.gz')
```
